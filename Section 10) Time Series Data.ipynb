# 시계열 데이터 분석을 위한 판다스 Expanding and Rolling 이해하기

```python
import pandas as pd
import numpy as np 
% matplotlib inline
```

```python
s = pd.Series(np.random.randn(1000), index = pd.date_range('1/1/2017',periods = 1000))
s.plot()
```

```python
s= s.cumsum()
s.plot()
```

```python
#window사이즈만큼 값을 뜀
r = s.rolling(window = 60)
#이동평균
r.mean()
```

```python
s.plot(style='k--')
```

```python
s.plot(style = 'k--')
r.mean().plot(style = 'k') #window 사이즈를 줄이면 줄일수록 두 그래프가 겹침
```

즉, rolling은 추세선을 만들때 사용

```python
df = pd.DataFrame(np.random.randn(1000,4), index = pd.date_range('1/1/2017',periods=1000), columns = ['A','B','C','D'])
df
```

```python
df = df.cumsum()
df.plot()
```

```python
df.rolling(window=30).sum().plot(subplots=True)
```

```python
df.rolling(window=len(df), min_periods= 1).mean().plot()
```

```python
df.expanding(min_periods = 1).mean().plot()
```

```python
dfe = pd.DataFrame({'B':[0,1,2,np.nan,4]})
dfe.plot()
```

```python
dfe.expanding(2).sum()
```

```python
dfe.expanding(2).sum().plot()
```

```python
dfe.expanding(2).mean().plot()
```
