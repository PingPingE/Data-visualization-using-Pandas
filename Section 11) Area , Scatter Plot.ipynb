```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
mpl.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.family'] = 'KBIZmjo M'
```

```python
df = pd.DataFrame(np.random.rand(10,4), columns = list('abcd'))
#NaN값은 저절로 0으로 찍기때문에 없애거나 채우는게 나음
#area plot은 값을 누적함(기본)
df.dropna().plot.area()
```

```python
df.plot()
```

```python
df.plot.area(stacked = False, grid=True)#누적 해제, grid
```

```python
df.plot.scatter(x = 'd', y= 'c',s = 50, grid = True)
```

```python
#두개 그리기
ax = df.plot.scatter(x = 'a' , y ='b', color = 'DarkBlue', label = 'Group 1')
df.plot.scatter(x = 'c', y= 'd', color = 'DarkGreen', label = 'Group 2' , ax = ax)
```

```python
df.plot.scatter(x = 'a', y= 'b', c = 'd' ,s = 50)#3개
```
