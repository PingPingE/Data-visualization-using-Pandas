```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
mpl.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.family'] = 'KBIZmjo M'
```

## kind : str
    - 'line' : line plot (default)
    - 'bar' : vertical bar plot
    - 'barh' : horizontal bar plot
    - 'hist' : histogram
    - 'box' : boxplot
    - 'kde' : Kernel Density Estimation plot
    - 'density' : same as 'kde'
    - 'area' : area plot
    - 'pie' : pie plot
    - 'scatter' : scatter plot
    - 'hexbin' : hexbin plot

```python
#시계열 데이터
ts = pd.Series(np.random.randn(1000), index = pd.date_range('1/1/2015', periods = 1000)) 
ts.plot()
```

```python
#누적합
ts = ts.cumsum()
ts.plot()
```

```python
#시계열 데이터
df = pd.DataFrame(np.random.randn(1000,4), index = ts.index, columns = list('ABCD'))
df.head()
```

```python
df = df.cumsum()
df.plot()
```

```python
df2 = pd.DataFrame(np.random.randn(1000,2), columns = list('BC'))
df2.head()
```

```python
df2['A'] = pd.Series(list(range(len(df2))))
df2.head()
df2 = df2.cumsum()
```

```python
df2.plot(x= 'A', y = 'B')
```

```python
df.iloc[5].plot(kind = 'bar')
#수평선 만들기
plt.axhline(0, color = 'k')
```

```python
df.plot.scatter(x = 'A', y = 'B')
```

```python
df3 = pd.DataFrame(np.random.randn(10, 4), columns = list('abcd'))

df3 = df3.cumsum()
df3.tail()
```

```python
df3.plot.bar()
```

```python
df3.plot.bar(stacked = True)#누적
```

```python
df3.plot.barh(stacked = True)
```

```python
df3['b'].diff().hist()#차분(f(x) - f(x-1))
```

```python
df3['a_diff'] = df3.a.diff()
df3['a_shift'] = df3.a.shift(1)#한칸 뒤로
df3['a_minus'] = df3['a'] - df3['a_shift']
df3[[ 'a','a_diff','a_shift', 'a_minus']].head()
```

```python
df3[['a','b','c']].diff().hist(color = 'k' , alpha = 0.5, bins = 10)#alpha는 투명도, bins 가 클수록 촘촘함
```
