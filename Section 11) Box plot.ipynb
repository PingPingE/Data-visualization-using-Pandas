# Box Plot
최댓값, 3사분위수, 2사분위수, 1시분위수, 최솟값을 모두 나타냄

```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
mpl.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.family'] = 'KBIZmjo M'
```

```python
#randn(생성 개수, 범위) rand(행,열)
df = pd.DataFrame(np.random.rand(10,5), columns = list('abcde'))
df
```

```python
#25%(1사분위수), 50%(중간값, 2사분위수), 75%(3사분위수)
df.describe()
```

```python
df.plot.box()
```

```python
#색 바꾸기
color = {'boxes' : 'DarkGreen', 'whiskers':'DarkOrange','medians':'DarkBlue','caps':'Gray'}
df.plot.box(color = color, sym = 'r+')
```

```python
df.plot.box(vert = False)#x,y축 바꾸기
```

```python
df2 = pd.DataFrame(np.random.rand(10,5))
df2
```

```python
bp = df2.boxplot()#grid추가
```

```python
df3 = pd.DataFrame(np.random.rand(10,2),columns = {'col1','col2'})
df3
                   
```

```python
#A와B로 그룹핑하기
df3['x']=pd.Series(['A','A','A','A','A','A','B','B','B','B','B','B'])
plt.figure();
bp = df3.boxplot(by='x')
```

```python
#seed값이 같으면 random 함수 써도 같은 숫자 나옴
np.random.seed(1234)
df_box = pd.DataFrame(np.random.randn(50,2))
df_box['g'] = np.random.choice(['A','B'],size = 50)
df_box.loc[df_box['g']=='B',1] += 3
bp = df_box.boxplot(by='g')
```
